# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
import statsmodels.api as sm
x = app_train['AMT_INCOME_TOTAL']
```

```python
ecdf = sm.distributions.ECDF(x)
plt.plot(ecdf.x, ecdf.y)
plt.show()
```

```python
x = app_train['AMT_INCOME_TOTAL']
def customized_EDCF(x):
    sort_value = np.sort(x)
    x_values = np.linspace(start = min(sort_value), stop = max(sort_value), num=10000)
    data_size = x.size # Number of elements in the array
    y_values = []
    
    for i in x_values:
        temp = x[x <= i]
        value = temp.size / data_size
        y_values.append(value)
    return x_values, y_values
```

```python
c_edcf = customized_EDCF(x)
plt.plot(c_edcf[0], c_edcf[1])
plt.show()
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
x = app_train['AMT_INCOME_TOTAL'] 
p_ecdf = sm.distributions.ECDF(x)
c_ecdf = customized_EDCF(x)

plt.plot(p_ecdf.x, p_ecdf.y)
plt.plot(c_edcf[0], c_edcf[1])
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([p_ecdf.x[1], p_ecdf.x.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
print(p_ecdf.x)
print(p_ecdf.x[1])
plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(p_ecdf.x),  p_ecdf.y)
plt.plot(np.log(c_ecdf[0]) , c_ecdf[1])
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
